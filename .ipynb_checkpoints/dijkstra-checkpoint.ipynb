```python
start = 'S'
end = 'E'

def lowestCost(costs, vistedNodes):
    cost = float("inf")
    nNode = None
    #for nextNode in [x for x in costs.keys() if x not in vistedNodes and x != end]:
    for nextNode in [x for x in costs.keys() if x not in vistedNodes]:
        if costs[nextNode] < cost:
            cost = costs[nextNode]
            nNode = nextNode
    return nNode


def dijkstra(graph, costs, parents):
    vistedNodes = []
    node = lowestCost(costs, vistedNodes)
    while node is not None:
        print("lowest node ", node)
        cost = costs[node]
        neighbors = graph.get(node, {}) 
        for (k,v) in neighbors.items():
            nCost = cost + neighbors[k]
            if costs[k] > nCost:
                costs[k] = nCost
                parents[k] = node
        print("costs: ", costs)
        print("parents: ", parents)
        vistedNodes.append(node)
        node = lowestCost(costs, vistedNodes)
    
if __name__ == "__main__":
    graph = {}
    graph['S'] = {'A':6, 'B':2, 'C':2}
    graph['A'] = {'E': 1}
    graph['B'] = {'A':3,'E':5}
    
    #graph['A'] = {'B':7, 'C':9, 'D':11}
    #graph['B'] = {'A':7, 'C': 8, 'F':20}
    #graph['C'] = {'A':9, 'B':8, 'E':6}
    #graph['D'] = {'A':11, 'E':3}
    #graph['E'] = {'C':6, 'D':3, 'F':1}
    #graph['F'] = {'B':20, 'E':1}

    parents =  {}
    costs = {}
    for node in graph.keys():
        for n in graph[node].keys():
            if node == start:
                costs[n] = graph[node][n]
                costs[node] = 0
                parents[n] = node
            elif costs.get(n) is None:
                costs[n] = float("inf")
            #parents[n] = node
    print(costs)
    dijkstra(graph, costs, parents)
    print(costs)
    print(parents)
    node = end
    while parents.get(node) is not None:
        print(node, "(", graph[parents.get(node)][node], ")<-", end="")
        node = parents.get(node)
    print(node)
        
```
